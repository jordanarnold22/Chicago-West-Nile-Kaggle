### Creating a dataframe that has the rate of westnile per station, in addition to the number of tests run at each station and the number of positives read at each station

```python
westnile = []
for i in range(1,151):
    x = combined.loc[combined['station'] == i]
    westnile.append(x['Predictions'].sum()/len(x['Predictions']))
```

```python
tested_density = []
for i in range(1,151):
    x = combined.loc[combined['station'] == i]
    tested_density.append(len(x['Predictions']))
```

```python
positives = []
for i in range(1,151):
    x = combined.loc[combined['station'] == i]
    positives.append(x['Predictions'].sum())
```

```python
where2spray = pd.DataFrame()
```

```python
where2spray['test_site'] = range(1,151)
```

```python
where2spray['tested_density'] = tested_density
```

```python
where2spray['positives'] = positives
```

```python
where2spray['westnile'] = westnile
```

```python
where2spray.head()
```

```python
where2spray2 = where2spray.sort_values(by=['westnile'],ascending=False)
```
