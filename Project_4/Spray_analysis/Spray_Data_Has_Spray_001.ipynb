### External reasearch approximates the porportion of 1 degree of lat an lon that is equal to 0.5 km in chicago.

.00625 for longitude in each direction

.0045 for latitude in each direction

### Creating a unique list of lat/lon pairs to make a testing station list

```python
train_t = pd.DataFrame(train['Latitude'])
```

```python
train_t['Longitude'] = train['Longitude']
```

```python
train_y = train_t.drop_duplicates()
```

```python
train_y = train_y.reset_index(drop=True)
```

### Left merging the dataframes so that the train set will have the station feature

```python
combined = pd.merge(train,train_y,on=['Latitude'])
```

```python
combined = combined.drop(columns=['Longitude_x','station_x'])
```

```python
combined['station'] = combined['station_y']
```

```python
combined = combined.drop(columns=['station_y'])
```

```python
combined['longitude'] = combined['Longitude_y']
```

```python
combined = combined.drop(columns=['Longitude_y'])
```

```python
combined.head()
```

### Creating a buffer zone around each station approximately equal to 1km

```python
combined['max_lon'] = combined['longitude'] + .00625
```

```python
combined['min_lon'] = combined['longitude'] - .00625
```

```python
combined['max_lat'] = combined['Latitude'] + .0045
```

```python
combined['min_lat'] = combined['Latitude'] - .0045
```

```python
combined.head()
```

### Checking how many unique days they sprayed on

```python
spray['Ordinal'].unique()
```

### Filtering my train dataset so that I my for loop later is more computationally efficent 

```python
combine_spray_dates = combined.loc[combined['Ordinal'].isin([734378, 734387, 735066, 735074, 735088, 735095, 735096, 735102,
       735109, 735116])] 
```

```python
combine_spray_dates.shape
```

```python
combine_spray_dates = combine_spray_dates.reset_index(drop=True)
```

```python
combine_spray_dates
```

### Creating the Has Spray feature by comparing the lat/lon pair in each spray row to the zones created in the training set.

```python
for rows in range(0,801):
    for index, row in spray.iterrows():
        if row['Ordinal'] == combine_spray_dates['Ordinal'][rows] and row['Latitude'] >= combine_spray_dates['min_lat'][rows] and row['Latitude'] <= combine_spray_dates['max_lat'][rows] and row['Longitude'] >= combine_spray_dates['min_lon'][rows] and row['Longitude'] <= combine_spray_dates['max_lon'][rows]:
            combine_spray_dates['Has_Spray'][rows] = 1
```

```python
combined['Has_Spray'].unique()
```

### Creating a sprayed_sites dataframe with only rows that have been sprayed

```python
sprayed_sites = combine_spray_dates.loc[combine_spray_dates['Has_Spray']==1]
```
