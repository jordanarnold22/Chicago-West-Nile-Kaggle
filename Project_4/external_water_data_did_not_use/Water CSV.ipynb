```python
url = 'https://nwis.waterdata.usgs.gov/il/nwis/measurements/?site_no=05536105&agency_cd=USGS'

res = requests.get(url)
```

```python
res.status_code
```

```python
soup = BeautifulSoup(res.text, 'lxml')
```

```python
table = soup.find('body').find('table', {'class' : 'tablesorter dataList'}).find('tbody')
water = pd.DataFrame(columns=range(0,13), index = range(0,183))
```

```python
row_marker = 183

for row in table.find_all('tr'):
    column_marker = 0
    columns = row.find_all('td')
    row_marker -= 1
    for column in columns:
        water.iat[row_marker,column_marker] = column.get_text()
        column_marker += 1
        
```

```python
water.columns = ['Measure Number', 'Date', 'Timezone', 'Measurement Used?', 'Who?', 'Measuring Agency',
                     'Stream Flow (ft3/s)', 'Gage Height (ft)', 'Gage Change (ft)', 'Meas. Duration (hr)',
                     'Meas. Rated', 'Control', 'Flow Adjust. Code']
```
