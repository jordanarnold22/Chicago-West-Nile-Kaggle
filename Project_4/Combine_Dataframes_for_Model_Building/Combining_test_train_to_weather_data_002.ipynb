### This notebook's purpose is to combine the weather data and the train / test data along the date column.

```python
import pandas as pd
pd.set_option('display.max_columns', 500)
```

### Importing Train Test and Weather CSV's

```python
weather = pd.read_csv('./weather_combined')
```

```python
train = pd.read_csv('./train_clean')
```

```python
test = pd.read_csv('./test_clean')
```

### Merging the weather and train CSV's on the Date Column

```python
combined = pd.merge(weather,train,on=['Date'])
combined = combined.drop(columns=['Unnamed: 0_x','Unnamed: 0_y'])
```

```python
combined.isnull().sum().sum()
```

```python
combined.head()
```

### Exporting to CSV

```python
combined.to_csv('merged_train_weather',index=False)
```

### Merging the Weather and Test CSV's on the Date Column

```python
combined_2 = pd.merge(weather,test,on=['Date'])
combined_2 = combined_2.drop(columns=['Unnamed: 0_x','Unnamed: 0_y'])
```

```python
combined_2.head()
```

### Export to CSV

```python
combined_2.to_csv('merged_test_weather',index=False)
```
