### Reading in Weather CSV's

```python
weather_1 = pd.read_csv('./eda/weather/station_1.csv')
```

```python
weather_1 = weather_1.drop(columns=['Unnamed: 0'])
```

```python
weather_1.head()
```

```python
weather_2 = pd.read_csv('./eda/weather/station_2.csv')
```

```python
weather_2 = weather_2.drop(columns=['Unnamed: 0'])
```

```python
weather_2.head()
```

```python
weather_1_temp = weather_1.drop(columns = ['Station', 'Date','BCFG', 'BR', 'CALM', 'DZ', 'FG', 'FG+', 'FU', 'GR', 'HZ', 'MIFG', 'RA',
       'SN', 'SQ', 'TS', 'TSRA', 'VCFG', 'VCTS'])
```

```python
weather_2_temp = weather_2.drop(columns = ['Station', 'Date','BCFG', 'BR', 'CALM', 'DZ', 'FG', 'FG+', 'FU', 'GR', 'HZ', 'MIFG', 'RA',
       'SN', 'SQ', 'TS', 'TSRA', 'VCFG', 'VCTS'])
```

### Concacting weather 1 and 2 together 

```python
df_concat = pd.concat((weather_1_temp, weather_2_temp))
```

```python
by_row_index = df_concat.groupby(df_concat.index)
df_means = by_row_index.mean()
```

```python
weather_1.columns
```

```python
weather_1_dummies = weather_1.drop(columns = ['Station','Tmax', 'Tmin', 'Tavg', 'Depart', 'DewPoint',
       'WetBulb', 'Heat', 'Cool', 'Sunrise', 'Sunset', 'PrecipTotal',
       'StnPressure', 'SeaLevel', 'ResultSpeed', 'ResultDir', 'AvgSpeed'])
```

```python
weather_combined = pd.concat([weather_1_dummies, df_means], axis=1)
```

```python
weather_combined.head()
```

### Exporting to CSV

```python
weather_combined.to_csv('weather_combined')
```
