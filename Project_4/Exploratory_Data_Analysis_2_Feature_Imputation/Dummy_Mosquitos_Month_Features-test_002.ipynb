### The purpose of this sheet is to update the main CSV's to dummy out the Mosquito Species and to add the 'Month' feature as categorical data.

```python
import pandas as pd
import os 
os.chdir('/Users/gil/desktop/notebooks/DSI-US-4/Project4/')
pd.set_option('display.max_columns', 500)
train = pd.read_csv('./merged_test_weather_final')
```

```python
train.shape
```

### Seperating the Date into a month column

```python
temp = pd.DataFrame(train['Date'])
```

```python
temp2 = temp.assign(date_list=temp.Date.str.split('-'))
```

```python
temp2.head()
```

```python
temp2['month'] = 1
```

```python
temp2.shape
```

```python
for i in range(0,116293):
    temp2['month'][i] = temp2['date_list'][i][1] 
```

```python
temp2.head()
```

### Making dummies out of the month column

```python
temp3 = temp2.drop(columns=['Date', 'date_list'])
```

```python
temp4 = pd.get_dummies(temp3)
```

```python
temp4['month']=temp4['month'].astype('str')
```

```python
temp4 = pd.get_dummies(temp4)
```

```python
temp4.head()
```

### Making dummies out of the species

```python
train.head()
```

```python
species = pd.DataFrame(train['Species'])
```

```python
species = pd.get_dummies(species)
```

### Combine my final dataframes back into my training set and drop the species column

```python
species.shape
```

```python
comb = pd.merge(train, species,left_index=True, right_index=True)
```

```python
comb2 = pd.merge(comb, temp4,left_index=True, right_index=True)
```

```python
comb2 = comb2.drop(columns=['Species'])
```

```python
comb2.head()
```

```python
comb2.to_csv('merged_test_weather-finalv2')
```
